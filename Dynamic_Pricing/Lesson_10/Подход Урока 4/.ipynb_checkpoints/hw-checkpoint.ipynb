```python
def get_num_unique_vendors_brands(group_sku_filename, sku_dict):
    group_sku_df = pd.read_csv(f'Домашнее задание 2 (урок 4)/{group_sku_filename}')
    print(f"{group_sku_filename} shape: {group_sku_df.shape}")
    
    joined_df = group_sku_df.merge(sku_dict, how='left', left_on='SKU', right_on='sku_id')
    print(f"{group_sku_filename}: joined_df shape: {joined_df.shape}")
    
    return group_sku_filename.split('.')[0], joined_df['vendor'].nunique(), joined_df['brand_code'].nunique()

unique_vendor_brand_records = []

for group_sku_filename in ['first_group_sku.csv', 'second_group_sku.csv', 'third_group_sku.csv']:
    unique_vendor_brand_records.append(get_num_unique_vendors_brands(group_sku_filename, sku_dict))
    
unique_vendor_brand_df = pd.DataFrame(data=unique_vendor_brand_records, 
                                      columns=['file', 'vendor_unique_number', 'brand_unique_number'])
unique_vendor_brand_df.to_csv('outputs/unique_vendor_brand.csv', index=False)
```

```python
def get_sku_min_max_prices(group_sku_filename):
    group_sku_df = pd.read_csv(f'Домашнее задание 2 (урок 4)/{group_sku_filename}')
    print(f"{group_sku_filename} shape: {group_sku_df.shape}")
    
    df = group_sku_df.groupby('SKU')['price_per_sku'].agg(['min', 'max'])

    return df.reset_index(drop=False).rename(columns={'min': 'min_price', 'max': 'max_price'})

sku_min_max_prices = pd.DataFrame(columns=['SKU', 'min_price', 'max_price'])

for group_sku_filename in ['first_group_sku.csv', 'second_group_sku.csv', 'third_group_sku.csv']:
    sku_min_max_prices = sku_min_max_prices.append(get_sku_min_max_prices(group_sku_filename))
    
sku_min_max_prices.to_csv('outputs/sku_min_max_prices.csv', index=False)
```

## Data Preparation

```python
def get_x_y(df, target_colname):
    return df.drop(columns=[target_colname]), df[target_colname]

def drop_na_target(x, y):
    non_null_target_inds = (~y.isna())
    return x[non_null_target_inds], y[non_null_target_inds]

def save_data(x=None, y=None, cv=None):
    if x is not None:
        joblib.dump(x, 'x.j')
    if y is not None:
        joblib.dump(y, 'y.j')
    if cv is not None:
        joblib.dump(cv, 'cv.j')

def plug_date_features(df, date_colname):
    df[date_colname] = pd.to_datetime(df[date_colname])
    df['year'] = df[date_colname].dt.year
    df['month'] = df[date_colname].dt.month
    df['day'] = df[date_colname].dt.day
    return df

def custom_cv_inds(x, date_colname, nfolds):
    custom_cv = []
    min_date = x[date_colname].min()
    max_date = x[date_colname].max()
    window_size = ((max_date - min_date) // nfolds).days
    window_size = pd.Timedelta(window_size, unit='D')
    current_date = min_date + window_size
    print(f"window size: {window_size}\n")
    while current_date < max_date:
        print(f'Current Date: {current_date}')
        train_inds = np.where(x[date_colname] < current_date)[0]

        if current_date + window_size >= max_date:
            print('Last Period Overflow: appending the remainder to the last val set')
            val_inds  = np.where(x[date_colname] >= current_date)[0]
        else:
            condition = (x[date_colname] > current_date) & (x[date_colname] < current_date + window_size)
            val_inds  = np.where(condition)[0]
        print(f'train_inds len: {len(train_inds)}; test_inds len: {len(val_inds)}\n')
        custom_cv.append((train_inds, val_inds))
        current_date += window_size
    return custom_cv
```

```python
def prepare_data(df, date_colname, drop_date=False, is_cv_needed=False, n_cv_folds=None):
    if 'num_purchases' in df.columns:
        df.drop(columns='num_purchases', inplace=True)
    df = plug_date_features(df, date_colname)
    if is_cv_needed:
        cv = custom_cv_inds(df, date_colname, n_cv_folds)
    else:
        cv = None
    if drop_date:
        df.drop(columns=date_colname, inplace=True)
    return df, cv
```

```python
x, y = get_x_y(all_group_sku, 'price_per_sku')
x, y = drop_na_target(x, y)
x, cv = prepare_data(x, 'dates', drop_date=True, is_cv_needed=True, n_cv_folds=10)
save_data(x,y,cv)
```
